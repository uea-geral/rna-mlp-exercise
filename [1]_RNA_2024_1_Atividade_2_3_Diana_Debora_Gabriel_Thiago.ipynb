```python
# Revertendo o one-hot-encoder aplicado nas colunas 'wilderness_area' e 'soil_type' e aplicando o Label Encoder nas mesmas

wilderness_df = [col for col in df.columns if 'wilderness_area' in col.lower()]
soil_df = [col for col in df.columns if 'soil_type' in col.lower()]

def dummies(df):
  dum_df = pd.get_dummies(df)
  from_dum = pd.from_dummies(dum_df)
  return from_dum

wild_new_df = dummies(df[wilderness_df]).add_prefix('Wilderness_Area')
soil_new_df = dummies(df[soil_df]).add_prefix('Soil_Type')
df_object = pd.concat([wild_new_df, soil_new_df], axis=1)

new_df = [i for i in df.columns if 'wilderness_area' not in i.lower() and 'soil_type' not in i.lower()]
dataframe = df[new_df]
data = pd.concat([dataframe, df_object], axis=1)

cols_objects = ['Wilderness_Area', 'Soil_Type']

for col in cols_objects:
    LE = LabelEncoder()
    data[col] = LE.fit_transform(data[col])
```

```python
# plotando o gráfico boxplot nas colunas preditoras do dataset para visualização da distribuição, tendência central e dispersão

# retirando o atributo alvo
data_X = data.drop('Cover_Type', axis=1)

# plotando o gráfico
fig, axs = plt.subplots(nrows=4, ncols=3, figsize=(18, 16))
ax = axs.flatten()
columns = data_X.columns
plt.suptitle("Boxplot das variáveis preditoras", fontsize=24)

for x, i in enumerate(columns):
    sns.boxplot(data_X[i], ax=ax[x])
    ax[x].set_title(i, fontsize=14)
    ax[x].set_ylabel('')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# plotando a distribuição da variável alvo
plt.figure(figsize=(8, 6))
sns.countplot(x='Cover_Type', data=data)
plt.title('Distribuição das classes da variável alvo', fontsize=16)
plt.xlabel('Classes', fontsize=14)
plt.ylabel('Frequência', fontsize=14)
plt.show()
```

###A partir da visualização elaborada, o que pode-se dizer a respeito do balanceamento do atributo-alvo?

- O gráficos mostram que diversas variáveis, como Elevation, Slope e Hillshade, apresentam padrões concentrados em faixas específicas, sugerindo similaridades entre as classes alvo, embora haja exceções. Enquanto isso, outras variáveis, como Aspect, exibem distribuições uniformes, indicando pouca variação entre as classes alvo. Algumas, como Horizontal_Distance_To_Hydrology, mostram uma ampla variação de valores e presença de outliers. Além disso, há uma forte associação entre as classes alvo e certas áreas selvagens e tipos de solo. No entanto, o estudo identifica um desbalanceamento significativo nas instâncias das classes alvo, com algumas sendo muito mais frequentes do que outras, o que pode afetar por exemplo, a performance dos modelos de machine learning.
